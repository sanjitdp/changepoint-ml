```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
alpha = 0.01
num_rounds = 1000
timeseries_length = 1000
changepoint = 200

e_processes = np.zeros(timeseries_length)
wealth = np.zeros(timeseries_length)

change_found = False
for t in range(timeseries_length):
    curr_obs = (
        np.random.normal(0, 1)
        if t < changepoint
        else np.exp(np.random.normal(0, 1)) - np.exp(-0.5)
    )

    e_processes[t] = 1
    e_processes += np.sign(curr_obs) * (e_processes > 0)

    wealth[t] = np.max(e_processes)
```

```python
plt.plot(wealth)
```
